# Extra Notebook - how to change the overall color of a texture map
This is another topic that will not be covered in detail.  If you want to make a planet that is outside of what you are given this notebook will take you through a simple example.

We are going to do our mods with an image library:

```python
import PIL.Image as Image
# if the above gives an error try:
#!conda install -c anaconda pillow
```

```python
im = Image.open('/Users/jillnaiman1/Downloads/planetFiles/textureMaps/jupiter_1200.jpg', 'r')
```

Let's plot it as a figure with matplotlib.

```python
import matplotlib.pyplot as plt
%matplotlib inline

fig,ax = plt.subplots()
ax.imshow(im)
```

Cool, that looks pretty cool.  Let's now turn this into an array so we can mess with it.  Let's first start by translating it into a numpy array and then turning up the red:

```python
import numpy as np

imarr = np.array(im)
imarr
```

The above are RGB triplets.  Let's make all the reds really red:

```python
imarr_red = imarr.copy()

imarr_red[:,:,0] = 255
```

Now we can show as an image again:

```python
plt.imshow(imarr_red)
```

Compared to the above, its a lot redder.  We can also do other tricks like turn down the red and up the green:

```python
imarr_green = imarr.copy()
```

Turn down the red and make sure its in an integer format:

```python
imarr_green[:,:,0] = (imarr_green[:,:,0]*0.5).astype('uint8')
```

Turn up the green and again ensure its in an integer format:

```python
imarr_green[:,:,2] = (imarr_green[:,:,2]*3.0).astype('uint8')
```

Make sure we don't go outside the 0-255 coding:

```python
imarr_green[:,:,2][imarr_green[:,:,2] > 255] = 255
```

Finally, lets show it:

```python
plt.imshow(imarr_green)
```

Once you are done with your image you can save it as a JPEG:

```python
imout = Image.fromarray(imarr_green)
imout.save('/Users/jillnaiman1/Downloads/planetFiles/textureMaps/jupiter_green.jpg')
```
