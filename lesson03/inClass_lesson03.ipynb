```python
# let's try to plot 3 lines over eachother
# x = stay the same
# y will be some randomly generated values -- I want this change each time
# on the 2nd loop, I want the plotted line to fancy!

for i in range(3): # this is a "counting" kind of loop up to 3
    #print(i)
    # *if* we are on the 2nd loop, or i is equal 1 then do something fancy
    # otherwise, just plot the default line
    y = np.random.random(100)
    if (i==1):
        # do a fancy plot
        plt.plot(x,y, '--', lw=5) # dashed line
    else:
        # do a default plot
        plt.plot(x,y)
    
plt.show()
```

## Let's talk about functions!

```python
print("Hey I'm using a function!") # print is the function, its called with (), inputs are a string
```

Steps to making a function:
 1. Think about it!  What are the things I want it do?  Do I want it to return something (calculation, word) -- Outputs? What does it need to do what I want it to? -- Inputs.  We'll probably write some pseudo-code
 2. Define the function: `def FunctionName(inputs)`
 3. Use the function -- "calling the function".  This is where many errors can pop up.

```python
# technically will work if I've run the cell below, but not a good idea!
#myPrint()
```

```python
# define a function that prints "hi there" three times

def myPrint(): # function definition statement: "def" is the defining key word, "myPrint" is the function name
               #                                 and it takes in no inputs -- ()
    for i in range(3): # print things three times --- for loop
        print("hi there")
    # no return statement
```

```python
# last thing to do: use the function:

myPrint()
```

```python
# redefine this function to take in as a *parameter* a changing number of times to print

def myPrintWithParameters(numberOfPrints): # numberOfPrints will control how many times the loop loops
    # note: all parameters used inside the function are only know to the function
    # I can't call "numberOfPrints" outside
    #print('The value of numberOfPrints is', numberOfPrints)
    for i in range(numberOfPrints):
        print("hi there")
```

```python
# call the function to print 5 times
myPrintWithParameters(5)
```

```python
#numberOfPrints # this gives us an error because numberOfPrints only "lives" inside our function
```

```python
# make a new printing function that returns number of prints!

def myPrintWithReturn(numberOfPrints): # still inputting number of prints
    for i in range(numberOfPrints):
        print("hi there")
    # this indent level is within the function NOT the for loop
    return numberOfPrints
```

```python
numPrints = myPrintWithReturn(7)
```

```python
numPrints
```

```python
# make a new printing function that returns TWICE the number of prints!

def myPrintWithReturn2(numberOfPrints): # still inputting number of prints
    for i in range(numberOfPrints):
        print("hi there")
    # this indent level is within the function NOT the for loop
    twicePrints = numberOfPrints*2 # twicePrints lives ONLY in the function
    return twicePrints # returning the value of this variable to the outside
```

```python
myTwicePrints = myPrintWithReturn2(3)
```

```python
myTwicePrints
```
